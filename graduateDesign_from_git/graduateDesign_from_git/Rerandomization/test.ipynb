```python
def findDiagonalOrder(mat):

    n = len(mat)
    items = []
    i, j = 0, 0
    dir = -1
    for num in range(n**2):
        items.append(mat[i][j])
        print(i, j)
        if (i+dir<0 and j-dir>n-1):
            i += 1
            dir *= -1
        elif (i+dir>n-1 and j-dir<0):
            j += 1
            dir *= -1
        elif (i+dir<0) or (i+dir>n-1):
            j += 1
            dir *= -1
        elif (j-dir<0) or (j-dir>n-1):
            i += 1
            dir *= -1
        else:
            i += dir
            j -= dir
    return items
```

```python

def strStr(haystack, needle):
    m, n = len(haystack), len(needle)
    i = j = 0
    N = buildNext(needle)
    while (i < m-1 and j < n-1):
        if (j<0 or haystack[i] == needle[j]):
            i += 1
            j += 1
        else:
            j = N[j]
    
    return i - j 

def buildNext(P):
    m = len(P)  # Get length of pattern
    N = [-1] * m  # Initialize next array
    
    j, t = 0, -1
    
    while j < m - 1:
        if t < 0 or P[j] == P[t]:
            j += 1
            t += 1
            N[j] = t
        else:
            t = N[t]
            
    return N

        
```

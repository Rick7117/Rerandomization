```python
import pandas as pd
import numpy as np
from functions_for_simulation import *
```

```python
## initializing

np.random.seed(10)

rowNum, colNum = (300, 100)

Xs = Xs_generator(rowNum, colNum)
# Xs = np.random.normal(0, 4, n)
Ws_1 = np.ones(rowNum) 
Ws_0 = np.zeros(rowNum)

Ys_1 = Ys_generator(Xs, Ws_1)
Ys_0 = Ys_generator(Xs, Ws_0)

tau = np.average(Ys_1['y']) - np.average(Ys_0['y'])
# # If we use the same eps, tau will be the constant 1.0 

print("The average treatment effect tau: {0:.3f}".format(tau))
```

```python
clarify_crition = {
    'x0': np.average(Xs['x0']),
    'x1': np.average(Xs['x1']),
    'x2': np.average(Xs['x2'])
}
# clarify_crition = Xs.mean().to_dict()

Ws_stratified = Ws_generator_stratified(Xs, clarify_crition)
print('Ws generated by stratified_randomization is {0}.'.format(Ws_stratified))

df_stratified = Ys_generator(Xs, Ws_stratified)
randomization_output_plot(df_stratified)
```

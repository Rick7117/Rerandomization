```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成服从正态分布的100个随机数
data = pd.DataFrame({
    'x1': np.random.normal(0, 100, 10000),
    'x2': np.random.beta(2, 5, 10000),
    'x3': np.random.beta(5, 2, 10000),
    'x4': np.random.normal(0, 400, 10000),
})
data.describe()
```

```python
fig, axs = plt.subplots(1, 4, figsize=(20, 3))
axs[0].hist(data['x1'], bins=40, color='grey')
axs[0].axvline(x=np.mean(data['x1']), color='r')
axs[0].axvline(x=np.median(data['x1']), color='g')
axs[1].hist(data['x2'], bins=40, color='grey')
axs[1].axvline(x=np.mean(data['x2']), color='r')
axs[1].axvline(x=np.median(data['x2']), color='g')
axs[2].hist(data['x3'], bins=40, color='grey')
axs[2].axvline(x=np.mean(data['x3']), color='r')
axs[2].axvline(x=np.median(data['x3']), color='g')
axs[3].hist(data['x4'], bins=40, color='grey')
axs[3].axvline(x=np.mean(data['x4']), color='r')
axs[3].axvline(x=np.median(data['x4']), color='g')
```

```python
fig, axs = plt.subplots(1, 4, figsize=(20, 3))
axs[0].boxplot(data['x1'])
axs[1].boxplot(data['x2'])
axs[2].boxplot(data['x3'])
axs[3].boxplot(data['x4'])
```
